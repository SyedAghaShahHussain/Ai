```python
import heapq

start = (1,2,3,
         4,0,6,
         7,5,8)

goal  = (1,2,3,
         4,5,6,
         7,8,0)

def print_board(s):
    for i in range(0, 9, 3):
        row = s[i:i+3]
        print(" ".join("_" if x == 0 else str(x) for x in row))
    print()

def h(state):  # Manhattan distance
    dist = 0
    for i, v in enumerate(state):
        if v == 0:
            continue
        gi = goal.index(v)
        r1, c1 = divmod(i, 3)
        r2, c2 = divmod(gi, 3)
        dist += abs(r1-r2) + abs(c1-c2)
    return dist

def neighbors(state):
    z = state.index(0)
    r, c = divmod(z, 3)
    res = []

    def swap(nz):
        a = list(state)
        a[z], a[nz] = a[nz], a[z]
        return tuple(a)

    if r > 0: res.append((swap(z-3), "Up"))
    if r < 2: res.append((swap(z+3), "Down"))
    if c > 0: res.append((swap(z-1), "Left"))
    if c < 2: res.append((swap(z+1), "Right"))

    return res

def astar():
    pq = []
    heapq.heappush(pq, (h(start), 0, start))  # (f, g, state)

    parent = {start: (None, None)}  # state -> (prev_state, move)
    g_cost = {start: 0}
    visited = set()

    while pq:
        f, g, cur = heapq.heappop(pq)

        if cur in visited:
            continue
        visited.add(cur)

        if cur == goal:
            # reconstruct path
            path = []
            s = cur
            while parent[s][0] is not None:
                prev, move = parent[s]
                path.append((move, s))
                s = prev
            path.reverse()

            print("START:")
            print_board(start)

            for i, (move, state) in enumerate(path, 1):
                print(f"Move {i}: {move}")
                print_board(state)

            print("✅ GOAL REACHED!")
            print("Goal State Values:")
            print_board(goal)
            print("Total Moves:", len(path))
            return

        for nxt, move in neighbors(cur):
            ng = g_cost[cur] + 1
            if nxt not in g_cost or ng < g_cost[nxt]:
                g_cost[nxt] = ng
                heapq.heappush(pq, (ng + h(nxt), ng, nxt))
                parent[nxt] = (cur, move)

    print("No solution found.")

astar()
```

```python
import heapq

graph = {
'A': [('B',1), ('C',3)],
'B': [('D',1), ('E',5)],
'C': [('F',2)],
'D': [('G',3)],
'E': [('G',2)],
'F': [('G',1)],
'G': []
}

heuristic = {
'A':7,
'B':6,
'C':4,
'D':3,
'E':2,
'F':1,
'G':0
}

def astar(start,goal):

    open_list=[]
    heapq.heappush(open_list,(0,start))

    g_cost={start:0}
    parent={}

    while open_list:

        current=heapq.heappop(open_list)[1]

        if current==goal:
            path=[]
            while current in parent:
                path.append(current)
                current=parent[current]

            path.append(start)
            path.reverse()
            return path

        for neighbor,cost in graph[current]:

            new_cost=g_cost[current]+cost

            if neighbor not in g_cost or new_cost<g_cost[neighbor]:

                g_cost[neighbor]=new_cost
                f=new_cost+heuristic[neighbor]

                heapq.heappush(open_list,(f,neighbor))
                parent[neighbor]=current

    return None


print("Shortest Path:",astar('A','G'))
```
